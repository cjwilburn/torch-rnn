```python
import agate
```

```python
dog_original = agate.Table.from_csv('data/dogs.txt')
```

```python
dog_sample = agate.Table.from_csv('data/dogs_sample_0.txt')
```

```python
distinct_original = dog_original.columns['dog_name'].values_distinct()
distinct_sample = dog_sample.columns[0].values_distinct()
```

```python
print(distinct_original)
```

```python
print(distinct_sample)
```

```python
# remove None types
originals = filter(None, distinct_original)
samples = filter(None, distinct_sample)
```

```python
# Convert all names to Titlecase
originals_titlcase = [x.title() for x in originals]
samples_titlecase = [x.title() for x in samples]
```

```python
print(samples_titlecase)
```

```python
for x in samples_titlecase:
    if x not in originals_titlcase:
        print(x)
```
